A vector is a list of numbers. We can think of a vector as a list of coordinates in space, or as an arrow from the origin to a point in space.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

These are just examples of **a Vector**. A vector is a list of numbers.

We can think of a vector as a point in space, with each element defining the distance along a different axis.
Where the initial point doesn't matter, only the distance along each axis. So a **vector is a direction and a magnitude**.

```python
# 2-dimensional vector
v2 = [ 3, -2 ]  # x, y

# plot them
plt.plot([0, v2[0]], [0, v2[1]])  # ([init_x, end_x], [init_y, end_y])
plt.axis('equal')  # make axes scales equal
plt.plot([-4, 4],[0, 0],'k--')  # plot x-axis
plt.plot([0, 0],[-4, 4],'k--')  # plot y-axis
plt.grid()  # make a grid
plt.axis((-4, 4, -4, 4))  # set x-axis and y-axis limits
plt.show()  # show the plot
```

```python
# 3-dimensional vector
v3 = [ 4, -3, 2 ]  # x, y, z
# row to column (or vice-versa):
v3t = np.transpose(v3)

# plot the 3D vector
fig = plt.figure(figsize=plt.figaspect(1))
ax = plt.axes(None, projection='3d')
ax.plot([0, v3[0]],[0, v3[1]],[0, v3[2]],linewidth=3)
ax.plot([2, 2+v3[0]],[1, 1+v3[1]],[2, 2+v3[2]],linewidth=3)  # same vector
ax.plot([4, 4+v3[0]],[2, 2+v3[1]],[4, 4+v3[2]],linewidth=3)  # same vector

# draw the axes
ax.plot([0, 0],[0, 0],[-4, 4],'k--')
ax.plot([0, 0],[-4, 4],[0, 0],'k--')
ax.plot([-4, 4],[0, 0],[0, 0],'k--')
plt.show()
```

## Operations on Vectors
To perform operations on vectors they must be of the same dimension (number of elements in the vector).

### Addition

$$\vec{a} + \vec{b} = \begin{bmatrix}a_1\\a_2\\a_3\end{bmatrix} + \begin{bmatrix}b_1\\b_2\\b_3\end{bmatrix} = \begin{bmatrix}a_1 + b_1\\a_2 + b_2\\a_3 + b_3\end{bmatrix}$$

### Subtraction

$$\vec{a} - \vec{b} = \begin{bmatrix}a_1\\a_2\\a_3\end{bmatrix} - \begin{bmatrix}b_1\\b_2\\b_3\end{bmatrix} = \begin{bmatrix}a_1 - b_1\\a_2 - b_2\\a_3 - b_3\end{bmatrix}$$

```python
# two vectors in R2
v1 = np.array([ 3, -1 ])
v2 = np.array([ 2,  4 ])
v3 = v1 + v2

# plot them
plt.plot([0, v1[0]],[0, v1[1]],'b',label='v1')
plt.plot([0, v2[0]]+v1[0],[0, v2[1]]+v1[1],'r',label='v2')
plt.plot([0, v3[0]],[0, v3[1]],'k',label='v1+v2')

plt.legend()
plt.axis('square')
plt.axis((-2, 6, -2, 6 ))
plt.grid()
plt.show()
```

### Scalar Multiplication

Scalar multiplication is the multiplication of a vector by a scalar (a real number). The **result is a vector with the same direction** as the original vector, but **with a magnitude that is the product of the original magnitude and the scalar**.

$$c\vec{a} = c\begin{bmatrix}a_1\\a_2\\a_3\end{bmatrix} = \begin{bmatrix}ca_1\\ca_2\\ca_3\end{bmatrix}$$

```python
# vector and scalar
v1 = np.array([ 3, -1 ])
l  = 4
v1m = v1*l # scalar-modulated

# plot them
plt.plot([0, v1[0]],[0, v1[1]],'b',label='$v_1$')
plt.plot([0, v1m[0]],[0, v1m[1]],'r:',label='$\lambda v_1$')

plt.legend()
plt.axis('square')
axlim = max([max(abs(v1)),max(abs(v1m))])*1.5 # dynamic axis lim
plt.axis((-axlim,axlim,-axlim,axlim))
plt.grid()
plt.show()
```

### Dot Product 
Also known as **Inner Product** or **Scalar Product** of two vectors $\vec{a}$ and $\vec{b}$.

$$\vec{a} \cdot \vec{b} = \begin{bmatrix}a_1\\a_2\\a_3\end{bmatrix} \cdot \begin{bmatrix}b_1\\b_2\\b_3\end{bmatrix} = \sum_{i=1}^n a_ib_i$$

The dot product is a scalar value. **It is also the product of the magnitudes of the two vectors and the cosine of the angle between them**.
$$\vec{a} \cdot \vec{b} = \|\vec{a}\| \|\vec{b}\| \cos \theta$$

```python
## many ways to compute the dot product

v1 = np.array([ 1, 2, 3,  4, 7, 6])
v2 = np.array([ 0,-4, 2, -3, 4, 5])

dp1 = sum(np.multiply(v1,v2))  # sum of element-wise multiplication
# dp4 = sum(map(lambda x,y: x*y, v1,v2))  # sum of element-wise multiplication (same as dp1 but is an slower implementation)

dp2 = np.dot(v1,v2)  # dot product
# dp5 = np.inner(v1,v2)  # inner product (same as dot product for 1D vectors)

dp3 = np.matmul(v1,v2)  # matrix multiplication

# print(dp1, dp2, dp3, dp4, dp5)
print(dp1, dp2, dp3)
```

$$\vec{a} \cdot \vec{b} > 0 \iff \cos{\theta} > 0$$

$$\vec{a} \cdot \vec{b} < 0 \iff \cos{\theta} < 0$$

$$\vec{a} \cdot \vec{b} = 0 \iff \cos{\theta} = 0 \iff \theta = 90^{\circ}$$

$$\|\vec{a}\| \|\vec{b}\| \iff \cos{\theta} = 1 \iff \theta = 0^{\circ}$$


```python
# two vectors
v1 = np.array([ 2,  4, -3 ])
v2 = np.array([ 0, -3, -3 ])

# dp6 = v1 length * v2 length * cos(theta)
ang = np.arccos( np.dot(v1,v2) / (np.linalg.norm(v1)*np.linalg.norm(v2)) )  # compute the angle (radians) between two vectors

fig = plt.figure()
fig = plt.figure(figsize=plt.figaspect(1))
ax = plt.axes(None, projection='3d')
ax.plot([0, v1[0]],[0, v1[1]],[0, v1[2]],'b')
ax.plot([0, v2[0]],[0, v2[1]],[0, v2[2]],'r')
plt.axis((-6, 6, -6, 6))
plt.title('Angle between vectors: %s rad.' %ang)
plt.show()

## equivalence of algebraic and geometric dot product formulas
print(f"algebraic {np.dot( v1,v2 )}")
print(f"geometric {np.linalg.norm(v1)*np.linalg.norm(v2)*np.cos(ang)}")
```

#### Distributive property

```python
n = 10
a = np.random.randn(n)
b = np.random.randn(n)
c = np.random.randn(n)

res1 = np.dot( a , (b+c) )
res2 = np.dot(a,b) + np.dot(a,c)

print([ res1,res2 ])
```

#### Associative Property
**For the dot product, the associative property isn't true**.

```python
n = 5
a = np.random.randn(n)
b = np.random.randn(n)
c = np.random.randn(n)

# the two results
res1 = np.dot( a , np.dot(b,c) )
res2 = np.dot( np.dot(a,b) , c )

# compare them
print(res1)
print(res2)

### special cases where associative property works!
# 1) one vector is the zeros vector
# 2) a==b==c
```

### Vector Length

The length of a vector is the square root of the sum of the squares of the vector elements.

$$\|\vec{a}\| = \sqrt{a_1^2 + a_2^2 + \cdots + a_n^2} = \sqrt{\sum_{i=1}^n a_i^2}$$

**Another way to see it** is to get the **square root of the dot product** of the vector with itself.

```python
# vector length
def vector_length(v):
    return np.sqrt(np.sum(v**2))

# vector normalization
def vector_normalization(v):
    return v / vector_length(v)


# test vector length
v = np.array([1, 2, 3])
print(vector_length(v))
print(np.linalg.norm(v))  # numpy built-in function (norm)
# the norm function can also be used to calculate the length of a vector

# plot vector in a 3d space
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(0, 0, 0, v[0], v[1], v[2], length=1, normalize=True)
# draw lines that represents the vector components
ax.plot([0, v[0]], [0, 0], [0, 0], color='r')
ax.plot([0, 0], [0, v[1]], [0, 0], color='g')
ax.plot([0, 0], [0, 0], [0, v[2]], color='b')

ax.set_xlim([0, 3])
ax.set_ylim([0, 3])
ax.set_zlim([0, 3])
# change the view angle
ax.view_init(azim=30)
ax.view_init(elev=30)
plt.show()
```

### Cross Product

$$\vec{a} \times \vec{b} = \begin{bmatrix}a_1\\a_2\\a_3\end{bmatrix} \times \begin{bmatrix}b_1\\b_2\\b_3\end{bmatrix} = \begin{bmatrix}a_2b_3 - a_3b_2\\a_3b_1 - a_1b_3\\a_1b_2 - a_2b_1\end{bmatrix}$$

### Outer Product

$$\vec{a} \otimes \vec{b} = \begin{bmatrix}a_1\\a_2\\a_3\end{bmatrix} \otimes \begin{bmatrix}b_1\\b_2\\b_3\end{bmatrix} = \begin{bmatrix}a_1b_1 & a_1b_2 & a_1b_3\\a_2b_1 & a_2b_2 & a_2b_3\\a_3b_1 & a_3b_2 & a_3b_3\end{bmatrix}$$
